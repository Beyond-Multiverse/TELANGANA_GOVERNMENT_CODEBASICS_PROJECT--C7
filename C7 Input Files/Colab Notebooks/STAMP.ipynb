```python
import csv
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_district = pd.read_csv("/content/drive/MyDrive/telangana/dim_districts.csv")
data_stamp = pd.read_csv("/content/drive/MyDrive/telangana/fact_stamps.csv")
```

```python
data_district
```

```python
data_stamp
```

```python
data_stamp['district'] = data_stamp['dist_code'].map(data_district.set_index('dist_code')['district'])
```

```python
data_stamp
```

```python
column_to_move = 'district'
new_order = [column_to_move] + [col for col in data_stamp.columns if col != column_to_move]
data_stamp = data_stamp[new_order]
data_stamp
```

```python
del data_stamp['dist_code']
```

```python
data_stamp
```

```python
data_stamp_2 = data_stamp.copy(deep=True)
```

```python
data_stamp_2
```

```python
data_stamp_2['month'] = pd.to_datetime(data_stamp_2['month'], format='%Y-%m-%d')
```

```python
data_stamp_2
```

```python
# Filter data for the two financial years
fy_2021 = data_stamp_2[(data_stamp_2['month'] >= '2021-04-01') & (data_stamp_2['month'] <= '2022-03-31')]
fy_2022 = data_stamp_2[(data_stamp_2['month'] >= '2022-04-01') & (data_stamp_2['month'] <= '2023-03-31')]
fy_2019 = data_stamp_2[(data_stamp_2['month'] >= '2019-04-01') & (data_stamp_2['month'] <= '2020-03-31')]
```

```python
fy_2022
```

```python
fy_2022_grouped = fy_2022.groupby('district').agg({
    'documents_registered_cnt': 'sum',
    'documents_registered_rev': 'sum',
    'estamps_challans_cnt': 'sum',
    'estamps_challans_rev': 'sum'
}).reset_index()
```

```python
fy_2022_grouped
```

```python
fy_2022_grouped['stamp_doc_ratio'] = ((fy_2022_grouped['estamps_challans_rev'] - fy_2022_grouped['documents_registered_rev']) / fy_2022_grouped['documents_registered_rev']) * 100
```

```python
fy_2022_grouped
```

```python
fy_2021_grouped = fy_2021.groupby('district').agg({
    'documents_registered_cnt': 'sum',
    'documents_registered_rev': 'sum',
    'estamps_challans_cnt': 'sum',
    'estamps_challans_rev': 'sum'
}).reset_index()
```

```python
fy_2021_grouped
```

```python
data_stamp_2
```

```python
cols_needed = ['district', 'documents_registered_rev']
q1a = data_stamp_2[cols_needed]
q1a
```

```python
q1a = q1a.groupby('district').agg({
    'documents_registered_rev': 'sum'
}).reset_index()
```

```python
q1a = q1a.sort_values(by='documents_registered_rev', ascending=False).reset_index(drop=True)
q1a
```

```python
fy_2019_grouped = fy_2019.groupby('district').agg({
    'documents_registered_cnt': 'sum',
    'documents_registered_rev': 'sum',
    'estamps_challans_cnt': 'sum',
    'estamps_challans_rev': 'sum'
}).reset_index()
```

```python
fy_2019_grouped
```

```python
columns_to_select = ['district', 'documents_registered_rev']
q1b = pd.merge(fy_2019_grouped[columns_to_select], fy_2022_grouped[columns_to_select], on='district', how='outer')
```

```python
q1b
```

```python
q1b = q1b.rename(columns={'documents_registered_rev_x': 'documents_registered_rev_2019'})
q1b = q1b.rename(columns={'documents_registered_rev_y': 'documents_registered_rev_2022'})
q1b
```

```python
q1b['documents_registered_rev_growth'] = ((q1b['documents_registered_rev_2022'] - q1b['documents_registered_rev_2019']) / q1b['documents_registered_rev_2019']) * 100
q1b
```

```python
q1b = q1b.sort_values(by='documents_registered_rev_growth', ascending=False).reset_index(drop=True)
```

```python
q1b = q1b.head(5)
q1b
```

```python
del q2a['estamps_challans_cnt']
del q2a['documents_registered_cnt']
q2a
```
