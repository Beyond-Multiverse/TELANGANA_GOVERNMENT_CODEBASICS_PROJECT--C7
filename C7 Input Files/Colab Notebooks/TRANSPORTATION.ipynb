```python
data_transport_2 = data_transport.copy(deep=True)
```

```python
data_transport_2
```

```python
selected_columns = ['district','month','fuel_type_petrol','fuel_type_diesel','fuel_type_electric']
data_transport_2 = data_transport_2[selected_columns]
```

```python
data_transport_2
```

```python
# Create a new DataFrame with the 'month' column converted to datetime
new_data_transport_2 = data_transport_2.copy()
new_data_transport_2['month'] = pd.to_datetime(data_transport_2['month'], format='%Y-%m-%d')

# Filter data for the two financial years
fy_2021 = new_data_transport_2[(new_data_transport_2['month'] >= '2021-04-01') & (new_data_transport_2['month'] <= '2022-03-31')]
fy_2022 = new_data_transport_2[(new_data_transport_2['month'] >= '2022-04-01') & (new_data_transport_2['month'] <= '2023-03-31')]

# Group by 'district' and sum the fuel types for each financial year
fy_2021_grouped = fy_2021.groupby('district').agg({
    'fuel_type_petrol': 'sum',
    'fuel_type_diesel': 'sum',
    'fuel_type_electric': 'sum'
}).reset_index()

fy_2022_grouped = fy_2022.groupby('district').agg({
    'fuel_type_petrol': 'sum',
    'fuel_type_diesel': 'sum',
    'fuel_type_electric': 'sum'
}).reset_index()

# Rename columns for each financial year
fy_2021_grouped.rename(columns={
    'fuel_type_petrol': 'petrol_21',
    'fuel_type_diesel': 'diesel_21',
    'fuel_type_electric': 'electric_21'
}, inplace=True)

fy_2022_grouped.rename(columns={
    'fuel_type_petrol': 'petrol_22',
    'fuel_type_diesel': 'diesel_22',
    'fuel_type_electric': 'electric_22'
}, inplace=True)

# Merge the two financial year DataFrames back together
final_df = fy_2021_grouped.merge(fy_2022_grouped, on='district', how='outer')
```

```python
# Calculate percentage change for each column
final_df['petrol_percentage_change'] = ((final_df['petrol_22'] - final_df['petrol_21']) / final_df['petrol_21']) * 100
final_df['diesel_percentage_change'] = ((final_df['diesel_22'] - final_df['diesel_21']) / final_df['diesel_21']) * 100
final_df['electric_percentage_change'] = ((final_df['electric_22'] - final_df['electric_21']) / final_df['electric_21']) * 100

```

```python
final_df
```

```python
# Select only the 'district' and percentage change columns
final_df = final_df[['district', 'petrol_percentage_change', 'diesel_percentage_change', 'electric_percentage_change']]

# Print the final DataFrame with the selected columns
final_df
```

```python
# Find the top 3 and bottom 3 districts for each percentage change column
top_petrol = final_df.nlargest(3, 'petrol_percentage_change')
bottom_petrol = final_df.nsmallest(3, 'petrol_percentage_change')

top_diesel = final_df.nlargest(3, 'diesel_percentage_change')
bottom_diesel = final_df.nsmallest(3, 'diesel_percentage_change')

top_electric = final_df.nlargest(3, 'electric_percentage_change')
bottom_electric = final_df.nsmallest(3, 'electric_percentage_change')

# Print the top and bottom districts for each percentage change column
print("Top 3 districts for Petrol Percentage Change:")
print(top_petrol)
print("\nBottom 3 districts for Petrol Percentage Change:")
print(bottom_petrol)

print("\nTop 3 districts for Diesel Percentage Change:")
print(top_diesel)
print("\nBottom 3 districts for Diesel Percentage Change:")
print(bottom_diesel)

print("\nTop 3 districts for Electric Percentage Change:")
print(top_electric)
print("\nBottom 3 districts for Electric Percentage Change:")
print(bottom_electric)
```
