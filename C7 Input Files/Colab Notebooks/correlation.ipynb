```python
import csv
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_stamp = pd.read_csv("/content/drive/MyDrive/telangana/fact_stamps.csv")
```

```python
data_stamp['district'] = data_stamp['dist_code'].map(data_district.set_index('dist_code')['district'])
column_to_move = 'district'
new_order = [column_to_move] + [col for col in data_stamp.columns if col != column_to_move]
data_stamp = data_stamp[new_order]
data_stamp
```

```python
del data_stamp['dist_code']
data_stamp
```

```python
data_pass = pd.read_csv("/content/drive/MyDrive/telangana/fact_TS_iPASS.csv")
```

```python
data_pass['district'] = data_pass['dist_code'].map(data_district.set_index('dist_code')['district'])
column_to_move = 'district'
new_order = [column_to_move] + [col for col in data_pass.columns if col != column_to_move]
data_pass = data_pass[new_order]
data_pass
```

```python
del data_pass['dist_code']
data_pass
```

```python
data_pass['month'] = pd.to_datetime(data_pass['month'], format='%d-%m-%Y')
```

```python
data_pass
```

```python
# Filter data for the two financial years
fy_2021_pass = data_pass[(data_pass['month'] >= '2021-04-01') & (data_pass['month'] <= '2022-03-31')]
fy_2022_pass = data_pass[(data_pass['month'] >= '2022-04-01') & (data_pass['month'] <= '2023-03-31')]
```

```python
# Filter data for the two financial years
fy_2021_stamp = data_stamp[(data_stamp['month'] >= '2021-04-01') & (data_stamp['month'] <= '2022-03-31')]
fy_2022_stamp = data_stamp[(data_stamp['month'] >= '2022-04-01') & (data_stamp['month'] <= '2023-03-31')]
```

```python
# Filter data for the two financial years
fy_2021_transport = data_transport[(data_transport['month'] >= '2021-04-01') & (data_transport['month'] <= '2022-03-31')]
fy_2022_transport = data_transport[(data_transport['month'] >= '2022-04-01') & (data_transport['month'] <= '2023-03-31')]
```

```python
fy_2021_stamp = fy_2021_stamp.groupby('district').agg({
    'estamps_challans_rev': 'sum'
}).reset_index()

fy_2021_stamp
```

```python
fy_2022_stamp = fy_2022_stamp.groupby('district').agg({
    'estamps_challans_rev': 'sum'
}).reset_index()

fy_2022_stamp
```

```python
t_electric_2022 = fy_2022_transport.groupby('district').agg({
    'fuel_type_electric': 'sum'
}).reset_index()

t_electric_2021 = fy_2021_transport.groupby('district').agg({
    'fuel_type_electric': 'sum'
}).reset_index()

t_agriculture_2022 = fy_2022_transport.groupby('district').agg({
    'vehicleClass_Agriculture': 'sum'
}).reset_index()

t_agriculture_2021 = fy_2021_transport.groupby('district').agg({
    'vehicleClass_Agriculture': 'sum'
}).reset_index()

t_agriculture_2021
```

```python
fy_2022_transport['total_vehicle'] = fy_2022_transport[['category_Non-Transport','category_Transport']].sum(axis=1)
fy_2021_transport['total_vehicle'] = fy_2021_transport[['category_Non-Transport', 'category_Transport']].sum(axis=1)

total_2022 = fy_2022_transport.groupby('district').agg({
    'total_vehicle': 'sum'
}).reset_index()

total_2021 = fy_2021_transport.groupby('district').agg({
    'total_vehicle': 'sum'
}).reset_index()

total_2021
```

```python
automobile_2021 = fy_2021_pass[(fy_2021_pass['sector'] == 'Automobile')]
automobile_2022 = fy_2022_pass[(fy_2022_pass['sector'] == 'Automobile')]
electric_2021 = fy_2021_pass[(fy_2021_pass['sector'] == 'Electrical and Electronic Products')]
electric_2022 = fy_2022_pass[(fy_2022_pass['sector'] == 'Electrical and Electronic Products')]
agriculture_2021 = fy_2021_pass[(fy_2021_pass['sector'] == 'Agro based incl Cold Storages') | (fy_2021_pass['sector'] == 'Fertlizers Organic and Inorganic,Pesticides,Insecticides, and Other Related')]
agriculture_2022 = fy_2022_pass[(fy_2022_pass['sector'] == 'Agro based incl Cold Storages') | (fy_2022_pass['sector'] == 'Fertlizers Organic and Inorganic,Pesticides,Insecticides, and Other Related')]
agriculture_2022
```

```python
automobile_2021 = automobile_2021.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

automobile_2022 = automobile_2022.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

electric_2021 = electric_2021.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

electric_2022 = electric_2022.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

agriculture_2021 = agriculture_2021.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

agriculture_2022 = agriculture_2022.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

totalpass_2021 = fy_2021_pass.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()

totalpass_2022 = fy_2022_pass.groupby('district').agg({
    'investment in cr': 'sum'
}).reset_index()
```

```python
automobile_2022
```

```python
fy_2022 = pd.merge(fy_2022_stamp, totalpass_2022, on='district', how='outer')
fy_2022 = pd.merge(fy_2022, total_2022, on='district', how='outer')
fy_2022
```

```python
fy_2021 = pd.merge(fy_2021_stamp, totalpass_2021, on='district', how='outer')
fy_2021 = pd.merge(fy_2021, total_2021, on='district', how='outer')
fy_2021
```

```python
fy22 = pd.merge(total_2022, automobile_2022, on='district', how='outer')
fy22 = pd.merge(fy22, electric_2022, on='district', how='outer')
fy22 = pd.merge(fy22, t_electric_2022, on='district', how='outer')
fy22 = pd.merge(fy22, agriculture_2022, on='district', how='outer')
fy22 = pd.merge(fy22, t_agriculture_2022, on='district', how='outer')

# Create a dictionary to map old column names to new column names
column_name_mapping = {'investment in cr_x': 'automobile_investment',
                       'investment in cr_y': 'electric_investment',
                       'investment in cr': 'agriculture_investment'}

fy22.rename(columns=column_name_mapping, inplace=True)
fy22
```

```python
fy21 = pd.merge(total_2021, automobile_2021, on='district', how='outer')
fy21 = pd.merge(fy21, electric_2021, on='district', how='outer')
fy21 = pd.merge(fy21, t_electric_2021, on='district', how='outer')
fy21 = pd.merge(fy21, agriculture_2021, on='district', how='outer')
fy21 = pd.merge(fy21, t_agriculture_2021, on='district', how='outer')

# Create a dictionary to map old column names to new column names
column_name_mapping = {'investment in cr_x': 'automobile_investment',
                       'investment in cr_y': 'electric_investment',
                       'investment in cr': 'agriculture_investment'}

fy21.rename(columns=column_name_mapping, inplace=True)
fy21
```
